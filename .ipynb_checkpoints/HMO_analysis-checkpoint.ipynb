```python
# Bode Lab, from Aline Frietas
oligo_table_df = pd.read_excel('bode-lab/SEPSIS_HMO_Report_Obs_20221201.xlsx', index_col=1, dtype=str, header=1)
oligo_metadata_df = pd.read_csv('bode-lab/SEPSIS_Metadata_Milk_HMO_Obs_SchedA_v1_20221004.csv', sep=',', index_col=0)
oligo_metadata_df = oligo_metadata_df.join(oligo_table_df.iloc[:,2:5])
oligo_metadata_df = oligo_metadata_df.join(oligo_table_df.iloc[:,25:28])
oligo_table_df.index.name = "#SampleID"
oligo_metadata_df.index.name = "#SampleID"
oligo_table_df.index = ['14834.' + str(idx) for idx in oligo_table_df.index]
oligo_metadata_df.index = ['14834.' + str(idx) for idx in oligo_metadata_df.index]
oligo_table_df_ft = oligo_table_df.iloc[:,5:24]
oligo_metadata_df = oligo_metadata_df.rename(columns = {"COLLECTION_TIMESTAMP": "sample_collection_date"})
```

```python
oligo_obs_table_df = oligo_table_df_ft.copy()
oligo_obs_md_df = oligo_metadata_df.copy()
```

```python
specific_cols = ["Secretor", "host_age", "host_age_infant", "Diversity", "Evenness", "Sia", "Fuc"]

merged_md_df = pd.concat([
    oligo_obs_md_df[specific_cols]
], axis=0, verify_integrity=True)

merged_ft_df = pd.concat([
    oligo_obs_table_df
], axis=0, verify_integrity=True)

merged_md_df = merged_md_df[merged_md_df['Secretor'] != '?']
merged_ft_df = merged_ft_df[merged_ft_df.index.isin(merged_md_df.index)]
```

```python
merged_df = pd.merge(merged_ft_df, merged_md_df[['Secretor', 'host_age_infant']], 
                    left_index=True, right_index=True)
merged_df['Secretor'] = merged_df['Secretor'].map({'1': 'True', '0': 'False'})
merged_df = merged_df.dropna(subset=['Secretor', 'host_age_infant'])

hmo_groups = {
    'Fucosylated': ['2\'FL', '3FL', 'DFLac', 'LNFP I', 'LNFP II', 'LNFP III', 'DFLNT', 'FLNH', 'DFLNH'],
    'Sialylated': ['3\'SL', '6\'SL', 'LSTb', 'LSTc', 'DSLNT', 'FDSLNH', 'DSLNH'],
    'Neutral': ['LNT', 'LNnT', 'LNH']
}

hmo_order = []
for group, hmos in hmo_groups.items():
    hmo_order.extend(hmos)

age_groups = {
    '< 30 days': merged_df[merged_df['host_age_infant'] < 30],
    '≥ 30 days': merged_df[merged_df['host_age_infant'] >= 30]
}

age_visual_settings = {
    '< 30 days': {'alpha': 0.5, 'marker': 's'},
    '≥ 30 days': {'alpha': 0.8, 'marker': 'o'}
}

hmo_palette = {'True': '#81B6A3', 'False': '#EA956F'}

def get_significance_label(p):
    if p < 0.001: return '***'
    elif p < 0.01: return '**'
    elif p < 0.05: return '*'
    else: return ''

def create_hmo_plot(data_df, age_group_name):
    data_melted = pd.melt(data_df.drop(columns=['host_age_infant']), 
                        id_vars=['Secretor'], var_name='HMO', value_name='Concentration')
    
    data_melted['Group'] = data_melted['HMO'].map({hmo: group for group, hmos in hmo_groups.items() for hmo in hmos})
    data_melted['Concentration'] = pd.to_numeric(data_melted['Concentration'], errors='coerce')
    data_melted['HMO'] = pd.Categorical(data_melted['HMO'], categories=hmo_order, ordered=True)
    
    data_melted['Secretor'] = pd.Categorical(data_melted['Secretor'], categories=['True', 'False'], ordered=True)
    
    stat_results = {}
    for hmo in hmo_order:
        secretor_data = data_melted[(data_melted['HMO'] == hmo) & (data_melted['Secretor'] == 'True')]['Concentration'].dropna()
        non_secretor_data = data_melted[(data_melted['HMO'] == hmo) & (data_melted['Secretor'] == 'False')]['Concentration'].dropna()
        
        if len(secretor_data) > 0 and len(non_secretor_data) > 0:
            try:
                stat, p = mannwhitneyu(secretor_data, non_secretor_data)
                stat_results[hmo] = {'p': p, 'label': get_significance_label(p)}
            except:
                stat_results[hmo] = {'p': 1.0, 'label': ''}
    
    fig = plt.figure(figsize=(14, 6))
    sns.set(style="ticks", font_scale=1.2)
    
    visual_settings = age_visual_settings[age_group_name]
    
    box_props = {'edgecolor': 'black', 'facecolor': 'none', 'linewidth': 1.5}
    whisker_props = {'linewidth': 1.5, 'color': 'black'}
    median_props = {'linewidth': 2, 'color': 'black'}
    cap_props = {'linewidth': 1.5, 'color': 'black'}
    
    ax = sns.boxplot(x='HMO', y='Concentration', hue='Secretor', data=data_melted, 
                    palette=hmo_palette, fliersize=0, linewidth=1.5, dodge=True,
                    width=0.8, boxprops=box_props, whiskerprops=whisker_props, 
                    medianprops=median_props, capprops=cap_props)
    
    sns.stripplot(x='HMO', y='Concentration', hue='Secretor', data=data_melted, 
                  marker=visual_settings['marker'], dodge=True, palette=hmo_palette,
                  size=5, alpha=visual_settings['alpha'], jitter=0.3, ax=ax)
    
    ax.set_yscale('log')
    plt.title(f'HMO Profile by Secretor Status - {age_group_name}', fontsize=20)
    plt.xlabel('Human Milk Oligosaccharides', fontsize=16)
    plt.ylabel('Concentration (nmol/mL)', fontsize=16)
    plt.xticks(rotation=45, ha='right')
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[:2], labels[:2], title='Secretor Status', 
              title_fontsize='13', fontsize='12')
    
    divider_positions = []
    cumulative_count = 0
    for group, hmos in hmo_groups.items():
        cumulative_count += len(hmos)
        if cumulative_count < len(hmo_order):
            divider_positions.append(cumulative_count - 0.5)
    
    for pos in divider_positions:
        plt.axvline(pos, color='gray', linestyle='--', alpha=0.7, linewidth=1.5)
    
    y_max_global = data_melted['Concentration'].max()
    plt.ylim(None, y_max_global * 3)
    
    for i, hmo in enumerate(hmo_order):
        if hmo in stat_results and stat_results[hmo]['label']:
            y_max_hmo = data_melted[data_melted['HMO'] == hmo]['Concentration'].max()
            bar_height = y_max_hmo * 1.5
            plt.plot([i-0.2, i+0.2], [bar_height, bar_height], color='black', linewidth=1.5)
            plt.text(i, bar_height*1.1, stat_results[hmo]['label'], ha='center', va='bottom', fontsize=12)
    
    ax2 = plt.twiny()
    
    group_ticks = []
    group_labels = []
    x_ticks = ax.get_xticks()
    
    start_idx = 0
    for group_name, hmos in hmo_groups.items():
        end_idx = start_idx + len(hmos) - 1
        midpoint = (x_ticks[start_idx] + x_ticks[end_idx]) / 2
        group_ticks.append(midpoint)
        group_labels.append(group_name)
        start_idx = end_idx + 1
    
    ax2.set_xlim(ax.get_xlim())
    ax2.set_xticks(group_ticks)
    ax2.set_xticklabels(group_labels, fontsize=14)
    ax2.tick_params(top=False)
    
    plt.tight_layout()
    plt.savefig(f"out/fig/hmo_secretor-profiles_{age_group_name.replace(' ', '_').replace('≥', 'ge')}.png", dpi=300)
    plt.savefig(f"out/fig/hmo_secretor-profiles_{age_group_name.replace(' ', '_').replace('≥', 'ge')}.svg", dpi=300)
    plt.show()
    return fig

plots = {}
for age_name, age_data in age_groups.items():
    plots[age_name] = create_hmo_plot(age_data, age_name)
    plt.close()
```

```python
merged_df = merged_md_df.copy()
for col in ['host_age_infant', 'Diversity', 'Evenness', 'Sia', 'Fuc']:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

merged_df = merged_df.dropna(subset=['host_age_infant', 'Diversity', 'Evenness', 'Sia', 'Fuc'])

merged_df['Secretor'] = merged_df['Secretor'].map({'1': 'True', '0': 'False'})
merged_df = merged_df.dropna(subset=['Secretor'])  # This will drop the '?' samples

merged_df['age_category'] = np.where(merged_df['host_age_infant'] < 30, '< 30 days', '≥ 30 days')

age_visual_settings = {
    '< 30 days': {'alpha': 0.5, 'marker': 's'},
    '≥ 30 days': {'alpha': 0.8, 'marker': 'o'}
}

hmo_palette = {'True': '#81B6A3', 'False': '#EA956F'}

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
sns.set(style="whitegrid", font_scale=1.2)
plt.rc('grid', linestyle='--', alpha=0.7)

def plot_metric(ax, x, y, hue, data, title):
    for age_cat, settings in age_visual_settings.items():
        subset = data[data['age_category'] == age_cat]
        sns.scatterplot(
            x=x, y=y, 
            hue=hue, 
            data=subset, 
            palette=hmo_palette, 
            ax=ax, 
            alpha=settings['alpha'],
            marker=settings['marker'],
            s=80 
        )
    
    for status in data[hue].unique():
        subset = data[data[hue] == status]
        sns.regplot(
            x=x, y=y, 
            data=subset, 
            scatter=False, 
            color=hmo_palette[status], 
            ax=ax, 
            order=2, 
            truncate=False
        )
    
    ax.set_title(title, fontsize=16)
    ax.set_xlabel("Infant Age (months)", fontsize=12)
    
    from matplotlib.lines import Line2D
    legend_elements = []
    for status in data[hue].unique():
        for age_cat, settings in age_visual_settings.items():
            legend_elements.append(
                Line2D([0], [0], marker=settings['marker'], color='w', 
                       markerfacecolor=hmo_palette[status], markersize=8, 
                       alpha=settings['alpha'], label=f"{status}, {age_cat}")
            )
    ax.legend(handles=legend_elements, title='Secretor Status, Infant Age', title_fontsize=10, loc='upper right', fontsize=8)
    ax.tick_params(axis='both', which='major', labelsize=10)

plot_metric(ax1, 'host_age_infant', 'Diversity', 'Secretor', merged_df, 'HMO Diversity vs Infant Age')
ax1.set_ylabel("HMO Diversity", fontsize=12)

plot_metric(ax2, 'host_age_infant', 'Sia', 'Secretor', merged_df, 'Sialylated HMOs vs Infant Age')
ax2.set_ylabel("Sialylated HMOs", fontsize=12)

plot_metric(ax3, 'host_age_infant', 'Fuc', 'Secretor', merged_df, 'Fucosylated HMOs vs Infant Age')
ax3.set_ylabel("Fucosylated HMOs", fontsize=12)

for ax in [ax1, ax2, ax3]:
    ax.set_ylim(bottom=0)

plt.tight_layout()
plt.savefig("out/fig/hmo_diversity.png", dpi=300)
plt.savefig("out/fig/hmo_diversity.svg", dpi=300)
plt.show()

for metric in ['Diversity', 'Evenness', 'Sia', 'Fuc']:
    print(f"\nCorrelations for {metric}:")
    for status in merged_df['Secretor'].unique():
        subset = merged_df[merged_df['Secretor'] == status]
        correlation = subset['host_age'].corr(subset[metric])
        print(f"Correlation coefficient between Infant Age and {metric} for {status}: {correlation:.2f}")
```

```python
sns.set(style="whitegrid", font_scale=1.2)
plt.rc('grid', linestyle='--', alpha=0.7)

base_colors = {'True': '#81B6A3', 'False': '#EA956F'}

merged_df = pd.merge(merged_ft_df, merged_md_df[['Secretor', 'host_age_infant']], left_index=True, right_index=True)
merged_df['Secretor Status'] = merged_df['Secretor'].map({'1': 'True', '0': 'False'})
merged_df = merged_df.dropna(subset=['Secretor Status'])

valid_samples = merged_df[merged_df['Secretor Status'].isin(['True', 'False'])]
feature_columns = [col for col in merged_ft_df.columns]
feature_table = valid_samples[feature_columns]

scaled_data = StandardScaler().fit_transform(feature_table)
euclidean_distances = pdist(scaled_data, metric='euclidean')
euclidean_distance_matrix = squareform(euclidean_distances)
dm = DistanceMatrix(euclidean_distance_matrix, ids=feature_table.index)

permanova_result = permanova(dm, valid_samples['Secretor Status'], permutations=9999)
print(f"PERMANOVA result: p = {permanova_result['p-value']:.4f}")

sorted_df = valid_samples.sort_values(['Secretor Status', 'host_age_infant'])
sorted_idx = sorted_df.index
reorder_indices = [dm.ids.index(id) for id in sorted_idx if id in dm.ids]
reordered_dm = DistanceMatrix(
    dm.data[reorder_indices][:, reorder_indices],
    ids=[dm.ids[i] for i in reorder_indices]
)

fig = plt.figure(figsize=(9, 8))

main_ax_position = [0.15, 0.15, 0.7, 0.7]  # [left, bottom, width, height]
ax_main = fig.add_axes(main_ax_position)

secretor_status_sorted = sorted_df['Secretor Status'].values
ages_sorted = sorted_df['host_age_infant'].values
n_samples = len(sorted_idx)

is_secretor = secretor_status_sorted == 'True'
norm_ages = np.zeros_like(ages_sorted, dtype=float)
true_ages = ages_sorted[is_secretor]
false_ages = ages_sorted[~is_secretor]

if len(true_ages) > 0:
    min_true, max_true = min(true_ages), max(true_ages)
    norm_ages[is_secretor] = (true_ages - min_true) / (max_true - min_true) if max_true > min_true else 0.5
    
if len(false_ages) > 0:
    min_false, max_false = min(false_ages), max(false_ages)
    norm_ages[~is_secretor] = (false_ages - min_false) / (max_false - min_false) if max_false > min_false else 0.5

secretor_colors = []
for i, (status, age_norm) in enumerate(zip(secretor_status_sorted, norm_ages)):
    if status == 'True':
        alpha = 0.4 + 0.6 * age_norm
        color = mcolors.to_rgba(base_colors['True'], alpha=alpha)
    else:
        alpha = 0.4 + 0.6 * age_norm
        color = mcolors.to_rgba(base_colors['False'], alpha=alpha)
    secretor_colors.append(color)

im = ax_main.imshow(reordered_dm.data, cmap='viridis', aspect='equal')
ax_main.set_title('Euclidean Distance Matrix', fontsize=14, pad=20)
ax_main.set_xticks([])
ax_main.set_yticks([])

cbar_ax = fig.add_axes([0.86, 0.15, 0.03, 0.7])
cbar = plt.colorbar(im, cax=cbar_ax)
cbar.set_label('Distance')

ax_left = fig.add_axes([0.17, 0.15, 0.01, 0.7])
ax_left.imshow([[c] for c in secretor_colors], aspect='auto')
ax_left.set_xticks([])
ax_left.set_yticks([])

ax_top = fig.add_axes([0.19, 0.86, 0.62, 0.01])
ax_top.imshow([secretor_colors], aspect='auto')
ax_top.set_xticks([])
ax_top.set_yticks([])

legend_ax = fig.add_axes([0.15, 0.05, 0.7, 0.05])
legend_ax.set_axis_off()

plt.savefig("out/fig/hmo_distmatrix.png", dpi=300, bbox_inches='tight')
plt.savefig("out/fig/hmo_distmatrix.svg", dpi=300, bbox_inches='tight')
plt.show()
```

```python
def run_permanova(feature_table, group_labels):
    numeric_feature_table = feature_table.apply(pd.to_numeric, errors='coerce')
    numeric_feature_table = numeric_feature_table.fillna(0)
    
    rel_abundance = numeric_feature_table.div(numeric_feature_table.sum(axis=1), axis=0)
    bc_distances = pdist(rel_abundance, metric='braycurtis')
    bc_distance_matrix = squareform(bc_distances)
    
    dm = DistanceMatrix(bc_distance_matrix, ids=feature_table.index)
    permanova_result = permanova(dm, group_labels, permutations=9999)
    
    return dm, permanova_result

def individual_hmo_tests(feature_table, group_labels):
    numeric_feature_table = feature_table.apply(pd.to_numeric, errors='coerce')
    
    numeric_feature_table = numeric_feature_table.fillna(0)
    
    group1_data = numeric_feature_table[group_labels == 'True']
    group2_data = numeric_feature_table[group_labels == 'False']
    
    results = []
    
    for hmo in numeric_feature_table.columns:
        stat, pval = mannwhitneyu(
            group1_data[hmo],
            group2_data[hmo],
            alternative='two-sided'
        )
        
        mean1 = group1_data[hmo].mean()
        mean2 = group2_data[hmo].mean()
        
        results.append({
            'HMO': hmo,
            'MWU_statistic': stat,
            'p_value': pval,
            'Secretor_mean': mean1,
            'NonSecretor_mean': mean2,
            'Higher_in': 'Secretor' if mean1 > mean2 else 'NonSecretor'
        })
    
    results_df = pd.DataFrame(results)
    
    reject, pvals_corrected, _, _ = multipletests(
        results_df['p_value'], 
        alpha=0.05, 
        method='holm'
    )
    
    results_df['p_adjusted'] = pvals_corrected
    results_df['significant'] = reject
    
    results_df = results_df.sort_values('p_adjusted')
    
    return results_df

merged_df['Age Group'] = merged_df['host_age_infant'].apply(lambda x: '< 30 days' if x < 30 else '≥ 30 days')
merged_df['Combined Group'] = merged_df['Secretor Status'] + ', ' + merged_df['Age Group']

table_oligo_df_tf = merged_ft_df[merged_ft_df.index.isin(merged_df.index)]

dm_secretor, permanova_secretor = run_permanova(table_oligo_df_tf, merged_df['Secretor Status'])
dm_age, permanova_age = run_permanova(table_oligo_df_tf, merged_df['Age Group'])
dm_combined, permanova_combined = run_permanova(table_oligo_df_tf, merged_df['Combined Group'])

individual_tests = individual_hmo_tests(table_oligo_df_tf, merged_df['Secretor Status'])

print("PERMANOVA Results:")
print(f"By Secretor Status - Pseudo-F: {permanova_secretor['test statistic']:.2f}, p-value: {permanova_secretor['p-value']:.4f}")
print(f"By Age Group - Pseudo-F: {permanova_age['test statistic']:.2f}, p-value: {permanova_age['p-value']:.4f}")
print(f"By Combined Group - Pseudo-F: {permanova_combined['test statistic']:.2f}, p-value: {permanova_combined['p-value']:.4f}")

print("\nSignificant individual HMO differences by Secretor Status:")
sig_results = individual_tests[individual_tests['significant']]
secretor_higher = sig_results[sig_results['Higher_in'] == 'Secretor']['HMO'].tolist()
nonsecretor_higher = sig_results[sig_results['Higher_in'] == 'NonSecretor']['HMO'].tolist()

print(f"Higher in Secretors: {', '.join(secretor_higher)}")
print(f"Higher in Non-Secretors: {', '.join(nonsecretor_higher)}")
```
