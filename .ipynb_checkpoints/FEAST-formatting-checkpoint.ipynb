```python
feast_sse = pd.read_csv('../../data/Microbiome/FEAST_source-tracker/feces_same_sources_early_unrar_matrix_source_contributions_matrix.txt',
                      sep='\t', index_col=0)
feast_sse.index = feast_sse.index.str.removeprefix('X')
feast_sse.columns = feast_sse.columns.str.removeprefix('X')
feast_sse.head()
#total for each row (sink) is 1, sinks are all 
```

```python
feast_sse_reform = feast_sse.stack().reset_index()
feast_sse_reform.rename(columns={'level_0':'comparison 1', 'level_1': 'comparison 2', 0:'value'},
                        inplace=True)
feast_sse_reformat = feast_sse_reform[feast_sse_reform['value'] != 0].copy()
feast_sse_reformat.shape
#nans already dropped
```

```python
feast_sse_reformat['ID1'] = feast_sse_reformat['comparison 1'].str.split("_").str[0]
feast_sse_reformat['ID2'] = feast_sse_reformat['comparison 2'].str.split("_").str[0]

feast_sse_reformat['sink'] = feast_sse_reformat['comparison 1'].str.split("_", n=1).str[1]
feast_sse_reformat['source'] = feast_sse_reformat['comparison 2'].str.split("_", n=1).str[1]

feast_sse_reformat['sample_type2_sink'] = feast_sse_reformat['sink'].str.split("-").str[0]
feast_sse_reformat['sample_type2_source'] = feast_sse_reformat['source'].str.split("-").str[0]

feast_sse_reformat['infant-age-bin_sink'] = feast_sse_reformat['sink'].str.split("-").str[1]
feast_sse_reformat['infant-age-bin_source'] = feast_sse_reformat['source'].str.split("-").str[1]

feast_sse_reformat['source'].fillna(feast_sse_reformat['ID2'], inplace=True)
feast_sse_reformat['sample_type2_source'].fillna(feast_sse_reformat['ID2'], inplace=True)
feast_sse_reformat['infant-age-bin_source'].fillna(feast_sse_reformat['ID2'], inplace=True)

family_id_map = dict(zip(metadata.index, metadata['family_id']))
feast_sse_reformat['family-id_source'] = feast_sse_reformat['ID2'].map(family_id_map)
feast_sse_reformat['family-id_sink'] = feast_sse_reformat['ID1'].map(family_id_map)

feast_sse_reformat.head()
```

```python
feast_sse_reformat.sample_type2_sink.unique()
```

```python
feast_sse_reformat.sample_type2_source.unique()
```

```python
feast_sse_reformat2 = feast_sse_reformat[['sample_type2_source', 
                                          'family-id_sink', 'value']].groupby(['sample_type2_source', 
                                                                                 'family-id_sink']).sum()
#adds all the portions from each unique site together by family_id from sink
feast_sse_reformat2
```

```python
feast_sse_reformat2.reset_index().sample_type2_source.unique()
```

```python
early_ss = feast_sse_reformat2.reset_index().groupby(['sample_type2_source']).mean()
early_ss.rename(columns={'value':'value_early'}, inplace=True)
early_ss
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

pivot_df = comp_ss_grm.pivot_table(index='variable', columns='sample_type2_source', values='value')

# Colors/Labels
labels = ['Human Milk', 'Adult Feces', 'Vaginal', 'Unknown']
colors = ['#BE98C7', '#B4D8D8', '#F69320', 'grey']
source_order = ['adult_breast_milk', 'adult_feces', 'adult_vaginal_mucus', 'Unknown']

bottom = np.zeros(len(pivot_df))

# Plotting the stacked bars
for i, source in enumerate(source_order):
    ax.bar(pivot_df.index, pivot_df[source], bottom=bottom,
           label=labels[i], color=colors[i])
    bottom += pivot_df[source] # Update bottom for the next stack

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.set_ylabel("Mean Contribution")
#ax.set_xlabel("Variable")

plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('../../figures/feast_ss_stackedbar_mean_ifeces-sink.svg', dpi=150, bbox_inches='tight')
plt.show()
```
